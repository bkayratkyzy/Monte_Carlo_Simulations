```python
#Pull in spy data to calculate 10 month moving average

spy_hist_df = api.alpha_vantage.historic_quotes('SPY', adjusted=True, output_format='pandas')
spy_hist_df.head()


```

```python

#Drop columns we do not need

spy_hist_df.drop(['1. open', '2. high', '3. low', '4. close', '6. volume', 
                  '7. dividend amount', '8. split coefficient'], axis=1, inplace=True)
spy_hist_df.head()


#Sort earliest to latest 
spy_hist_df.sort_index(inplace=True, ascending=True)
spy_hist_df.head()

# Drop nulls
spy_hist_df.dropna(inplace=True)
spy_hist_df.head()

#Change column name
columns = ['SPY close']
spy_hist_df.columns = columns
spy_hist_df.head()


```

```python
#Calculate 10 month moving average

spy_hist_df['SMA 300'] = spy_hist_df['SPY close'].rolling(window=300).mean()

spy_hist_df.tail()
```

```python
#concat these two columns with returns_df

returns_df = pd.concat([
     spy_hist_df, returns_df], 
     axis = "columns", join = "inner")
returns_df.head()
```

```python
# Drop nulls
returns_df.dropna(inplace=True)
returns_df.head()
```

```python
returns_df['Signal'] = returns_df["SPY close"] - returns_df["SMA 300"]
returns_df.head()

#df['equal_or_lower_than_4?'] = df['set_of_numbers'].apply(lambda x: 'True' if x <= 4 else 'False')

```

```python
returns_df['BUY_SELL'] = returns_df['Signal'].apply(lambda x: 'BUY' if x >=0 else 'SELL')
returns_df.tail()
```

```python
#Strategy

#Once a month buy the top three performing sector etfs if BUY_SELL column = BUY, otherwise hold cash
```
